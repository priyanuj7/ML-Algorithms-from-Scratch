```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from collections import Counter
```

### KNN From Scratch

```python
class Knn:
    def __init__(self,k=5):
        self.n_neighbors = k
        self.X_train = None
        self.y_train = None
        self.X_test = None
        
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    
    
    def calculate_distance(self, point_a, point_b):
        return np.linalg.norm(point_a - point_b)
    
    def majority_count(self, neighbors):
        votes = []
        for i in neighbors:
            #since i here is not a number and a list containing the distance as well as the index therefore
            #using i[0] to get the index value from the list
            votes.append(self.y_train[i[0]])
        votes = Counter(votes)
        return votes.most_common()[0][0]
        
        
    def predict(self, X_test):
#         self.X_test = X_test
        y_pred = []
        for i in X_test:
            distances = []
            #calculate distance with each training point
            for j in self.X_train:
                distances.append(self.calculate_distance(i,j))
            #what enumerate does is that it brings in the index position as well along with the distance value
            top_n_neighbors = sorted(list(enumerate(distances)), key = lambda x:x[1])[0:self.n_neighbors]
            label = self.majority_count(top_n_neighbors)
            y_pred.append(label)
            
        return np.array(y_pred)
                
```

```python
clf_knn = Knn()
clf_knn.fit(X_train, y_train)
```

```python
y_pred_new = clf_knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_new)
```
